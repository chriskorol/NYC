```python
# Step 1: Create unified column
df_all['RateCode'] = df_all['RateCodeID'].fillna(df_all['RatecodeID'])

# Step 2: Drop both old versions
df_all = df_all.drop(columns=['RateCodeID', 'RatecodeID'])

# Step 3: Confirm it's clean
print("Missing in unified column:", df_all['RateCode'].isnull().sum())
```

```python
# Removing invalid data 
# Step 1: Remove Trips With Zero or Negative Distance

print("Trips with trip_distance <= 0:", (df_all['trip_distance'] <= 0).sum())

# Remove them
df_all = df_all[df_all['trip_distance'] > 0]
```

```python
# Step 2: Remove Trips With Zero or Negative Fare Amount

print("Trips with fare_amount <= 0:", (df_all['fare_amount'] <= 0).sum())

# Remove them
df_all = df_all[df_all['fare_amount'] > 0]
```

```python
# Step 3: Remove Trips With Zero Passengers 

print("Trips with passenger_count <= 0:", (df_all['passenger_count'] <= 0).sum())

# Remove them
df_all = df_all[df_all['passenger_count'] > 0]

```

```python
# Step 4: Remove Negative Tip or Total Amount

print("Trips with tip_amount < 0:", (df_all['tip_amount'] < 0).sum())
print("Trips with total_amount <= 0:", (df_all['total_amount'] <= 0).sum())

# Remove them
df_all = df_all[df_all['tip_amount'] >= 0]
df_all = df_all[df_all['total_amount'] > 0]
```

```python
# 1. Clean TRIP DISTANCE
# Keep trips between 0.1 and 100 miles
df_all = df_all[(df_all['trip_distance'] > 0.1) & (df_all['trip_distance'] <= 100)]

# 2. Clean FARE AMOUNT
# Keep realistic fares only
df_all = df_all[(df_all['fare_amount'] > 0) & (df_all['fare_amount'] <= 500)]

# 3. Clean PASSENGER COUNT
# Keep only 1–6 passengers
df_all = df_all[(df_all['passenger_count'] > 0) & (df_all['passenger_count'] <= 6)]

# 4. Clean LATITUDE & LONGITUDE (NYC bounding box)
df_all = df_all[
    (df_all['pickup_latitude'].between(40, 42)) &
    (df_all['pickup_longitude'].between(-75, -72)) &
    (df_all['dropoff_latitude'].between(40, 42)) &
    (df_all['dropoff_longitude'].between(-75, -72))
]

# 5. Remove NEGATIVE VALUES in financial columns
money_cols = [
    'extra', 'mta_tax', 'tip_amount', 
    'tolls_amount', 'improvement_surcharge', 'total_amount'
]

for col in money_cols:
    df_all = df_all[df_all[col] >= 0]

# Final check
print("Cleaned dataset shape:", df_all.shape)
```

```python
# 6. Feature Engineering (Hour, day, month)
#  Extract from pickup datetime
df_all['pickup_hour'] = df_all['tpep_pickup_datetime'].dt.hour
df_all['pickup_weekday'] = df_all['tpep_pickup_datetime'].dt.weekday  # 0 = Monday
df_all['pickup_month'] = df_all['tpep_pickup_datetime'].dt.month
df_all['pickup_day'] = df_all['tpep_pickup_datetime'].dt.day

# Add trip duration in minutes
df_all['trip_duration_min'] = (df_all['tpep_dropoff_datetime'] - df_all['tpep_pickup_datetime']).dt.total_seconds() / 60

# Optional: add weekend flag
df_all['is_weekend'] = df_all['pickup_weekday'].isin([5, 6])  # Saturday=5, Sunday=6
```

```python
import geopandas as gpd

# Ensure pickup points are a GeoDataFrame with geometry
gdf_pickup = gpd.GeoDataFrame(
    df_all,
    geometry=gpd.points_from_xy(df_all['pickup_longitude'], df_all['pickup_latitude']),
    crs="EPSG:4326"
)

# Spatial join to attach neighborhood polygons
gdf_joined = gpd.sjoin(gdf_pickup, gdf_neighborhoods, how="left", predicate="within")

# Save neighborhood names into df_all
df_all['pickup_neighborhood'] = gdf_joined['ntaname']


```

```python
import statsmodels.api as sm
import pandas as pd

results = {}

for neighborhood in df_all['pickup_neighborhood'].dropna().unique():
    sub_df = df_all[df_all['pickup_neighborhood'] == neighborhood].dropna(subset=['fare_amount', 'trip_distance'])

    if len(sub_df) > 50:
        X = sm.add_constant(sub_df[['trip_distance']])
        y = sub_df['fare_amount']

        model = sm.OLS(y, X).fit()

        results[neighborhood] = {
            'k': model.params['trip_distance'],
            'b': model.params['const'],
            'r2': model.rsquared
        }

# Convert to DataFrame
df_results = pd.DataFrame(results).T.sort_values(by='r2', ascending=False)
```

```python
# Step 1: Merge k-values into neighborhood GeoDataFrame
gdf_results = gdf_neighborhoods.copy()
gdf_results = gdf_results.set_index("ntaname")
gdf_results = gdf_results.join(df_results[['k']])
gdf_results = gdf_results.dropna(subset=["k"])

# Step 2: Convert GeoDataFrame to GeoJSON
geojson_data = gdf_results.reset_index().to_json()

# Step 3: Plot with folium
import folium

m = folium.Map(location=[40.75, -73.97], zoom_start=11)

folium.Choropleth(
    geo_data=geojson_data,
    name='Fare Rate (k)',
    data=gdf_results.reset_index(),
    columns=['ntaname', 'k'],
    key_on='feature.properties.ntaname',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.3,
    legend_name='Fare Rate per km ($)',
).add_to(m)

m.save("fare_rate_by_neighborhood.html")
```
